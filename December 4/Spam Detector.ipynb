```python
from tkinter import *
import joblib
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
def clean_text(sent):
    tokens1=word_tokenize(sent.lower())
    tokens2=[token for token in tokens1 if token.isalpha()]
    tokens3=[ps.stem(token) for token in tokens2 if token not in swords]
    return tokens3
```

```python
classifier = joblib.load("classifier.model")
preprocessor = joblib.load("preprocessor.model")
ps = PorterStemmer()
swords = stopwords.words("english")
```

```python
def check():
    message = t.get("1.0", END)
    mes_new = preprocessor.transform([message])
    prediction = classifier.predict(mes_new)
    result.config(text="Message is: " + str(prediction[0]), fg="green" if prediction[0] == 'ham' else "red")
```

```python
top = Tk()
top.title("Spam Detector")
top.geometry("600x300")
Label(top, text="Enter your message here: ").place(x=100, y=10)
t = Text(top, width=50, height=5)
t.place(x=100, y=40)
b = Button(top, text="Check Message", command=check)
b.place(x=100, y=130)
result = Label(top, text="Message is", borderwidth=2, font=("Arial", 20))
result.place(x=100, y=180)
top.mainloop()
```
