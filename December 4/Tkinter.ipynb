```python
def find():
    num = int(e.get())
    print("Square: ", num ** 2)
```

```python
top = Tk()
top.title("Hello World")
top.geometry("600x400")
l = Label(top, text="Enter number")
e = Entry()
b = Button(top, text="Find square", command=find)

l.pack()
e.pack()
b.pack()
top.mainloop()
```

```python
def find():
    num = int(e.get())
    result.config(text="Square is: " + str(num ** 2))


top = Tk()
top.title("Hello World")
top.geometry("600x400")
l = Label(top, text="Enter number")
e = Entry()
b = Button(top, text="Find square", command=find)
result = Label(top, text="Square is: ")

l.place(x=50, y=50)
e.place(x=50, y=80)
b.place(x=50, y=110)
result.place(x=50, y=140)

top.mainloop()
```

```python
def add():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text="Result: " + str(num1 + num2))


def sub():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text="Result: " + str(num1 - num2))


def mul():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text="Result: " + str(num1 * num2))


def div():
    num1 = int(e1.get())
    num2 = int(e2.get())
    result.config(text="Result: " + str(num1 / num2))


top = Tk()
top.title("Calculator")
top.geometry("400x300")
l1 = Label(top, text="Enter number1: ")
e1 = Entry()
l2 = Label(top, text="Enter number2: ")
e2 = Entry()
add_but = Button(top, text="+", command=add, padx=5, pady=5)
sub_but = Button(top, text="-", command=sub, padx=5, pady=5)
mul_but = Button(top, text="*", command=mul, padx=5, pady=5)
div_but = Button(top, text="/", command=div, padx=5, pady=5)
result = Label(top, text="Result: ")

l1.place(x=50, y=50)
e1.place(x=150, y=50)
l2.place(x=50, y=80)
e2.place(x=150, y=80)

add_but.place(x=50, y=120)
sub_but.place(x=80, y=120)
mul_but.place(x=110, y=120)
div_but.place(x=140, y=120)
result.place(x=50, y=170)
top.mainloop()
```
