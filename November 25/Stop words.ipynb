```python
from nltk.tokenize import word_tokenize
```

```python
tokens=word_tokenize(sent)
```

```python
clean=[w for w in tokens if w not in string.punctuation]
```

```python
print(clean)
```

```python
new_sent=''
for char in sent:
    if char.isalpha() or char.isdigit() or char==' ':
        new_sent=new_sent+char
```

```python
 from nltk.corpus import stopwords
 swords=stopwords.words('english')
```

```python
swords
```

```python
for word in new_sent.split():
    if word in swords:
        continue
    else:
        print(word)
```
