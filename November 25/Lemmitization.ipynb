```python
words1="consultant","consult","consultants","consulting","consultative"
words2='management',"managing","managed","manage"
words3="big","bigger","biggest"
words4="go","gone","going","went"
words5="association","associating","associated","associations"
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wnl=WordNetLemmatizer()
```

```python
for word in words4:
    print(wnl.lemmatize(word,pos='v'))
```

```python
for word in words1:
    print(wnl.lemmatize(word,pos='n'))
```

```python
for word in words2:
    print(wnl.lemmatize(word,pos='v'))
```

```python
for word in words5:
    print(wnl.lemmatize(word,pos='v'))
```

```python
for word in words1:
    print(wnl.lemmatize(word,pos='n'))
```

```python
for word in words3:
    print(wnl.lemmatize(word,pos='a'))
```

```python
print(wnl.lemmatize('is',pos='v'))
print(wnl.lemmatize('worst',pos='a'))
print(wnl.lemmatize('children',pos='n'))
print(wnl.lemmatize('worst',pos='a'))
```
