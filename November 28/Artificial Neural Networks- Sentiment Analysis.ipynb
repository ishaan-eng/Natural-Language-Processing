```python
import string, re, nltk, os
from nltk.corpus import stopwords
from tensorflow.keras.preprocessing.text import Tokenizer
import numpy as np
```

Load the doc in memory

```python
def load_doc(filename):
    f = open(filename)
    text = f.read()
    f.close()
    return text
```

```python
from nltk.tokenize import word_tokenize

swords = stopwords.words('english')
```

Function to clean the doc

```python
def clean_doc(doc):
    tokens = word_tokenize(doc)
    tokens = [token for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if token not in swords]
    tokens = [token for token in tokens if len(token) > 1]
    return tokens
```

Load all the files from directory

```python
def process_train(directory):
    documents = []
    for filename in os.listdir(directory):
        if not filename.startswith('cv9'):
            path = directory + '\\' + filename
            doc = load_doc(path)
            tokens = clean_doc(doc)
            documents.append(tokens)
    return documents
```

```python
def process_test(directory):
    documents = []
    for filename in os.listdir(directory):
        if filename.startswith('cv9'):
            path = directory + '\\' + filename
            doc = load_doc(path)
            tokens = clean_doc(doc)
            documents.append(tokens)
    return documents
```

```python
te = process_test(r'review_polarity\txt_sentoken\pos')
len(te)
```

```python
def process_docs(directory, is_train: bool) -> list:
    documents = []
    for filename in os.listdir(directory):
        if is_train and filename.startswith('cv9'):
            continue
        if not is_train and not filename.startswith('cv9'):
            continue
        path = directory + '\\' + filename
        doc = load_doc(path)
        tokens = clean_doc(doc)
        documents.append(tokens)
    return documents
```

```python
tr = process_docs(r'review_polarity\txt_sentoken\pos', True)
len(tr)
```

```python
te = process_docs(r'review_polarity\txt_sentoken\pos', False)
len(te)
```

Load data

```python
def load_data(is_train):
    neg = process_docs('review_polarity\\txt_sentoken\\neg', is_train)
    pos = process_docs('review_polarity\\txt_sentoken\\pos', is_train)
    docs = neg + pos
    labels = [0 for i in range(len(neg))] + [1 for i in range(len(pos))]
    return docs, labels
```

Load training data

```python
train_data, train_labels = load_data(True)
```

Loading testing data

```python
test_data, test_labels = load_data(False)
```

```python
len(train_data), len(train_labels)
```

Data Preparation

```python
def create_tokenizer(lines):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(lines)
    return tokenizer
```

```python
tokenizer = create_tokenizer(train_data)
```

Encode the training data

```python
X_train = tokenizer.texts_to_matrix(train_data)
```

```python
X_train.shape
```

Encode the testing data

```python
X_test = tokenizer.texts_to_matrix(test_data)
```

```python
X_test.shape
```

Build Model

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Embedding, Input
```

```python
def define_model(n_words):
    model = Sequential()
    model.add(Input(shape=(n_words,)))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = define_model(36389)
```

```python
from keras.utils import plot_model

plot_model(model, show_layer_activations=True, show_layer_names=True, show_shapes=True)
```

Train the model

```python
model.fit(X_train, np.array(train_labels), epochs=10, batch_size=10)
```

```python
model.evaluate(X_test, np.array(test_labels), batch_size=1)
```

```python
def predict_sentiment(review):
    clean_review = clean_doc(review)
    review_trns = tokenizer.texts_to_matrix([clean_review])
    pred = model.predict(review_trns)
    prob = pred[0,0]
    if prob > 0.5:
        return "Positive", prob
    return "Negative", 1-prob
```
