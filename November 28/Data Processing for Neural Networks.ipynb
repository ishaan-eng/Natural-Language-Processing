```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

import os
```

```python
lines = [
    'It was a nice rainy day.',
    'The things were so beautiful in his point.',
    'When your focus is clear, you won.',
    'Many many happy returns of the day.'
]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines)
```

```python
tokenizer.word_docs
```

```python
print(tokenizer.index_word)
```

```python
len(tokenizer.index_docs)
```

```python
tokenizer.word_index
```

```python
mat = tokenizer.texts_to_matrix(lines)
mat
```

```python
mat.shape
```

```python
seq = tokenizer.texts_to_sequences(lines)
```

```python
seq
```

```python
padded = pad_sequences(seq, maxlen=max([len(i) for i in seq]), padding='post')
```
