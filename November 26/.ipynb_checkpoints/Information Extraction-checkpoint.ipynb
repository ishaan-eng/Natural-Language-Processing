```python
import urllib
import urllib.request
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import MWETokenizer
```

```python
# url = urllib.request.urlopen("https://en.wikipedia.org/wiki/Amazon_rainforest")
```

```python
url = "https://en.wikipedia.org/wiki/Amazon_rainforest"
```

```python
headers = {
    'User-Agent': 'Custom-Script/1.9',
    'Accept-Language': 'en-US,en;q=0.5'
}
```

```python
response = requests.get(url, headers=headers)
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup.find("title")
```

```python
text = soup.get_text()
```

```python
print(text)
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
tokens = " ".join(tokens)
```

```python
tokens
```

```python
import string
from nltk import FreqDist
from nltk.corpus import stopwords
ls = word_tokenize(text)

tk1 = [t.lower() for t in ls if t.isalpha() or t.isdigit()]
tk2 = [t for t in tk1 if t not in stopwords.words('english')]
# final = [word for word in ls if not (word in stopwords.words('english') or word in string.punctuation)]
freq = FreqDist(tk2)
freq.plot(20);
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
tk3 = [ps.stem(t) for t in tk2]
freq = FreqDist(tk2)
freq.plot(20);
```

```python
url = "https://en.wikipedia.org/wiki/Tiger"
headers = {
    'User-Agent': 'Custom-Script/1.9',
    'Accept-Language': 'en-US,en;q=0.5'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text()
ls = word_tokenize(text)
tk1 = [t.lower() for t in ls if t.isalpha() or t.isdigit()]
tk2 = [t for t in tk1 if t not in stopwords.words('english')]
tk3 = [ps.stem(t) for t in tk2]
# final = [word for word in word_tokenize(text) if (word.isalpha() or word.isdigit()) and (word not in stopwords.words('english'))]
freq = FreqDist(tk3)
freq.plot(20);
```

```python
url = "https://en.wikipedia.org/wiki/Sachin_Tendulkar"
headers = {
    'User-Agent': 'Custom-Script/1.9',
    'Accept-Language': 'en-US,en;q=0.5'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
text = soup.get_text()
ls = word_tokenize(text)
mwe = MWETokenizer(mwes=[("Sachin", "Tendulkar")], separator=" ")
ls = mwe.tokenize(ls)
# print(ls)
tk1 = [t.lower() for t in ls if t.isalpha() or t.isdigit()]
tk2 = [t for t in tk1 if t not in stopwords.words('english')]
tk3 = [t for t in tk2]
# final = [word for word in word_tokenize(text) if (word.isalpha() or word.isdigit()) and (word not in stopwords.words('english'))]
freq = FreqDist(tk3)
freq.plot(20);
```

```python
freq.most_common(10)
```
