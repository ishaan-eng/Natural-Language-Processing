#### Find all meanings of a word

```python
word = str(input("Enter word: "))
for wrd in wordnet.synsets(word):
    print(f"Word: {wrd.lemmas()[0].name()} \n Definition: {wrd.definition()}")
```

#### Find all noun meanings

```python
word = str(input("Enter word: "))
for wrd in wordnet.synsets(word, pos='n'):
    print(f"Word: {wrd.lemmas()[0].name()} \n Definition: {wrd.definition()}")
```

#### Extract specific POS from the word

```python
verbs = dict()
nouns = dict()
adtjs = dict()
word = str(input("Enter word: "))
for wrd in wordnet.synsets(word):
    if wrd.pos() == 'n':
        nouns[wrd.lemmas()[0].name()] = wrd.definition()
    elif wrd.pos() == 'v':
        verbs[wrd.lemmas()[0].name()] = wrd.definition()
    elif wrd.pos() == 's':
        adtjs[wrd.lemmas()[0].name()] = wrd.definition()
    # print(f"Word: {wrd.lemmas()[0].name()} \n Definition: {wrd.definition()}")

nouns, verbs, adtjs
```

Retrieve the antonyms

```python
antonyms=set()

for word in wordnet.synsets("possible"):
    for lemma in word.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())
antonyms
```

```python
antonyms=set()

for word in wordnet.synsets("window"):
    for lemma in word.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())
antonyms
```

Extract the hypernyms

```python
word='soil'
```

```python
synsets=wordnet.synsets(word)
```

```python
for x in synsets:
    hypernyms=x.hypernyms()
    for h in hypernyms:
        print(word,'hypernym of ', h.name())
```

Extract Meronyms

```python
word='car'
synsets=wordnet.synsets(word)

for x in synsets:
    meronyms=x.part_meronyms()
    for m in meronyms:
        print(word,'meronym of ', m.name())
```

Extract Holonyms

```python
word='keyboard'
```

```python
synsets=wordnet.synsets(word)
for x in synsets:
    holonyms=x.part_holonyms()
    for m in holonyms:
        print(word,'holonym of ', m.name())
```
