```python
from nltk.tokenize import word_tokenize
from nltk.wsd import lesk
```

```python
sent1  = "This device is used to jam the signal."
sent2 = "There is heavy traffic jam here. Vehicles everywhere."
sent2 = "This jam is infuriating! We are stuck"
sent3 = "I love eating jam."
```

```python
from nltk.corpus import wordnet

verbs = dict()
nouns = dict()
adtjs = dict()
word = str(input("Enter word: "))
for wrd in wordnet.synsets(word):
    print(f"Word: {wrd.lemmas()[0].name()} \n Definition: {wrd.definition()}")

nouns, verbs, adtjs
```

```python
a1 = lesk(word_tokenize(sent1), "jam")
a1.definition()
```
