```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import ktrain
from ktrain import text
```

```python
df=pd.read_csv('SMSSpamCollection',sep='\t',names=['label','message'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
data_train,data_test=train_test_split(df,test_size=0.2,random_state=0)
```

```python
data_train.shape,data_test.shape
```

```python
(X_train,y_train),(X_test,y_test),preproc=text.texts_from_df(train_df=data_train,
                                                             text_column='message',
                                                             label_columns='label',
                                                             val_df=data_test,
                                                             maxlen=400,
                                                             preprocess_mode='bert')
```

```python
model=text.text_classifier(name='bert',
                          train_data=(X_train,y_train),preproc=preproc)
```

```python
learner=ktrain.get_learner(model=model,train_data=(X_train,y_train),val_data=(X_test,y_test),batch_size=16)
```

```python
predictor=ktrain.get_predictor(learner.model,preproc=preproc)
```
