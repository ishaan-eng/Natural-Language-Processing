```python
import nltk
import random

from jupyter_core.version import pattern
```

Define chatbot responses and intents

```python
responses={
    'greeting':['Hello!','Hi there!','Hey!','Howdy!'],
    'goodbye':['Goodbye!','Bye!','See ya!','Take Care!'],
    'thanks':['You\'re welcome!','Anytime!','Happy to assist!','Glad to help!'],
    'default':['Soory, I didnt understand.','Can you tell again?','I am not sure.']
}
```

define simple set of questions and responses or patterns

```python
pattern={
    'hi':'greeting',
    'hey':'greeting',
    'hello':'greeting',
    'howdy':'greeting',
    'bye':'goodbye',
    'thanks':'thanks',
    'thank you':'thanks',
}
```

```python
def respond(user_input):
    for word in nltk.tokenize.word_tokenize(user_input.lower()):
        if word in pattern :
            intent=pattern[word]
            return random.choice(responses[intent])
    return random.choice(responses['default'])
```

```python
respond('hi')
```

```python
respond('hi')
```

```python
def chat():
    print('Chatbot: How can I help you?(Type "exit" to quit)')
    while True:
        user_input = input("You: ")
        if user_input == 'exit':
            print('Chatbot: Goodbye')
            break
        response=respond(user_input)
        print('Chatbot: ', response)
```

```python
chat()
```
