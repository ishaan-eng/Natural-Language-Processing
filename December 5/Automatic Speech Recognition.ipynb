TTS

```python
import pyttsx3
```

```python
engine=pyttsx3.init()
```

Set the properties

```python
engine.setProperty('rate',150)
engine.setProperty('volume',0.9)
```

```python
text='Welcome to the world of Natural Language Processing using Python Programming'
```

```python
engine.say(text)
engine.runAndWait()
```
