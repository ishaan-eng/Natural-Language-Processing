```python
sent1='It is good practice for us.'
sent2='It was also good to know about it'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
new_text = tfidf.fit_transform([sent1, sent2])
```

```python
new_text.toarray()
```

```python
df=pd.DataFrame(new_text.toarray(),columns=tfidf.get_feature_names_out(),index=[sent1,sent2])
df
```

```python
with open('sample_sentences.txt', 'r') as f:
    sentences = [sen.strip() for sen in f.readlines()]
    new_text = tfidf.fit_transform(sentences)
    df=pd.DataFrame(new_text.toarray(),columns=tfidf.get_feature_names_out(),index=sentences)
df
```
